### 题目：输入数值n，按顺序打印从1到最大的n位数，例如输入n=3，则从1，2，3，一直打印到999

#### 陷阱：若使用循环遍历 1- 999...9 并依次输出，当位数n过大时，无论将其存入int或long或long long都会溢出，故使用字符串来模拟数字加法

```python
def Print1ToMaxOfNDigits(n):
    assert n > 0
    number = '0' * n
    isOver = False
    while isOver == False:
        isOver, number = Increment(number)
        if isOver == False: # 这句是防止满足最终进位还进行打印，最后一位将变为0
            PrintNumber(number)

# 每次对number加 1
def Increment(number):
    isOver = False
    carry = 0 # 是否进位
    nSum = 0 # 当前位的值
    for i in range(len(number) - 1, -1, -1): # 注意边界条件，右边界要写-1
        nSum = ord(number[i]) - ord('0') + carry # 有进位则加上,注意python字符之间无法直接相减，需要借用ord()
        if i == len(number) - 1: # 当循环在个位数层的时候，每次+1；当循环不在个位数层的时候，只需要处理进位
            nSum += 1 # 每调用一次Increment，这句也只会被执行一次，因为每次 Increment只是对number +1
        if nSum >= 10: #需要进位了
            if i == 0: # 如果已经到最高位了还要进位证明已经到了所有位数都已经是9了
                isOver = True
            nSum -= 10 # 当前位归零
            carry += 1 # 进位加一
            numList = list(number) # Python中的字符是无法改变的，得先变成list
            numList[i] = '0'
            number = ''.join(numList)          
        else:
            numList = list(number) # Python中的字符是无法改变的，得先变成list
            numList[i] = chr(nSum + ord('0'))
            number = ''.join(numList)
            break
    return isOver, number

def PrintNumber(number):
    # 去寻找第一个非'0'数字
    begin = 0
    for begin in range(len(number)):
        if number[begin] != '0':
            break
    for i in range(begin, len(number)):
        print(number[i], end = ' ')
    print()

Print1ToMaxOfNDigits(2)
```

### 易错点：
1. python中字符之间无法相减，ascii也无法直接变成字符。ord()将字符转为ascii，chr()将ascii转为字符
2. python的字符串是无法直接改变的，需要用list()先转换为列表，修改列表，再用''.join()变回字符
3. python函数传参传的不是地址，因此无法通过函数去改变一个值，只能用返回值的方式去改变
4. !False这样的写法是错误的
5. if isOver == False:                 # 这句是防止满足最终进位还进行打印，否则最后一位将变为0
              PrintNumber(number)

### 用数组的方式

```python
def Print1ToMaxOfNDigits(n):
    assert n > 0
    number = [0] * n
    isOver = False
    while isOver == False:
        isOver, number = Increment(number)
        if isOver == False:
            PrintNumber(number)

# 每次对number加 1
def Increment(number):
    isOver = False
    carry = 0 # 是否进位
    nSum = 0 # 当前位的值
    for i in range(len(number) - 1, -1, -1): # 注意边界条件，右边界要写-1
        nSum = number[i] + carry # 有进位则进位
        if i == len(number) - 1:
            nSum += 1
        if nSum >= 10:
            if i == 0:
                isOver = True
            nSum -= 10
            carry += 1
            number[i] = 0
        else:
            number[i] = nSum
            break
                    
    return isOver, number

def PrintNumber(number):
    # 去寻找第一个非 0 数字
    begin = 0
    for begin in range(len(number)):
        if number[begin] != 0:
            break
    for i in range(begin, len(number)):
        print(number[i], end = ' ')
    print()

Print1ToMaxOfNDigits(2)
```

# 全排列

```python
def Print1ToMaxOfNDigits3(n):
    if n <= 0:
        return 0
    number = [0] * n
    Print1ToMaxOfNDigitsRecursively(number, n, 0)

def Print1ToMaxOfNDigitsRecursively(number, length, index):
    if index == length:
        PrintNumber(number)
        return
    for i in range(10):
        number[index] = i
        Print1ToMaxOfNDigitsRecursively(number, length, index + 1)
        
Print1ToMaxOfNDigits3(2)
```
